```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
# Create a file path for the CSVs
summer_path= '/../Resources/summer.csv'
winter_path= '/../Resources/winter.csv'
dictionary_path= '/../Resources/dictionary.csv'
```

```python
# Turn the Summer CSV into a Dataframe

summer_df=pd.read_csv(summer_path)
summer_df

```

```python
# Turn the Winter CSV into a Dataframe
winter_df=pd.read_csv(winter_path)
winter_df
```

```python
# Turn the Dictionary
# CSV into a Dataframe
dictionary_df=pd.read_csv(dictionary_path)
dictionary_df
```

```python
# Merge the three dataframes into one

merged_df = pd.merge(summer_df, winter_df, how= 'outer')
merged_df
```

```python
# Rename the Country column to Code in order to match the other dataframes
merged_df = merged_df.rename(columns={'Country': 'Code'})
```

```python
# Merge the Dictionary Dataframe with the medal data and create an index on the Athlete values
final_merged_df = pd.merge(merged_df, dictionary_df, how= 'left', on= "Code")
final_merged_df.set_index("Athlete")


final_merged_df
```

```python
# Filter the data to the year 2014 to match the Population and GDP per Capita data

df_2014=final_merged_df.loc[final_merged_df['Year']== 2014]
df_2014
```

```python
# Create a dataframe of only gold medals

gold_df=final_merged_df.loc[final_merged_df['Medal'].str.contains("Gold", case=True)]


gold_df
```

```python
# Create a dataframe of only silver medals
silver_df=final_merged_df.loc[final_merged_df['Medal'].str.contains("Silver", case=True)]
silver_df
```

```python
# Create a dataframe of only bronze medals
bronze_df=final_merged_df.loc[final_merged_df['Medal'].str.contains("Bronze", case=True)]
bronze_df
```

```python
# Sort the data by country
gold_sort_df=gold_df.sort_values('Country', ascending=True,)

gold_sort_df
```

```python
# Use groupby to count the number of gold medals won by each country 

gold_count = gold_sort_df.groupby(['Code'])['Code'].count()
print(gold_count)
```

```python
# Use groupby to count the number of silver medals won by each country 

silver_count = silver_df.groupby(['Code'])['Code'].count()
print(silver_count)
```

```python
# Use groupby to count the number of bronze medals won by each country 

bronze_count = bronze_df.groupby(['Code'])['Code'].count()
print(bronze_count)
```

```python
# Create a dataframe containing the medal counts

medal_df = pd.DataFrame({"Total Gold Medals": gold_count,
                            "Total Silver Medals": silver_count,
                            "Total Bronze Medals": bronze_count
                               })
medal_df.head()





```
